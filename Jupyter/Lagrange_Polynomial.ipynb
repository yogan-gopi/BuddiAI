```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(-50, 50, size = (10))
x = np.sort(x)
y = [i**3 for i in x]

x, y
```

```python
plt.plot(x, y)
```

```python
### lAGRANGE'S fUNCTION

xplt = np.linspace(x[0], x[-1])
yplt = np.array([], float)

for xp in xplt:
    yp = 0
    for xi, yi in zip(x, y):
        yp += yi * np.prod((xp - x[x != xi])/(xi - x[x!= xi]))
    yplt = np.append(yplt, yp)
    
```

```python
plt.plot(x, y, 'ro', xplt, yplt, 'b-')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
