```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
x = np.random.randint(10, size = (100))
x
```

```python
x = np.random.randint(-100, 10, size = (10))
x = numpy.sort(x)
y = np.exp(x)

x, y
```

```python
plt.plot(x, y)
```

```python
xplt = np.linspace(x[0], x[-1])
yplt = np.array([], float)

for xp in xplt:
    yp = 0
    for xi, yi in zip(x, y):
        yp += yi * np.prod((xp - x[x != xi])/(xi - x[x!= xi]))
    yplt = np.append(yplt, yp)
```

```python
plt.plot(x, y, 'ro', xplt, yplt, 'b-')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
